```python
plt.scatter(dataset["volatile acidity"],dataset["fixed acidity"])
plt.title("A graph of volatile acidity against fixed acidity")
plt.xlabel("Volatile acidity")
plt.ylabel("Fixed acidity")
plt.show()
```

```python
plt.scatter(dataset["pH"],dataset["fixed acidity"])
plt.title("A graph of pH against fixed acidity")
plt.xlabel("pH")
plt.ylabel("Fixed acidity")
plt.show()
```

```python
plt.scatter(dataset["quality"],dataset["residual sugar"])
plt.title("A graph of quality against residual sugar")
plt.xlabel("quality")
plt.ylabel("residual sugar")
plt.show()
```

```python
plt.scatter(dataset["alcohol"],dataset["pH"])
plt.title("A graph of alcohol against pH")
plt.xlabel("Alcohol")
plt.ylabel("pH")
plt.show()
```
